
# Analyze trends in track duration over the years

```python
print("Columns in the dataset:")
print(spotify_data.columns)
```

```python
import matplotlib.pyplot as plt

# Plot a histogram of track durations
plt.figure(figsize=(10, 6))
plt.hist(spotify_data['Duration (Minutes)'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Track Durations')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
